```python
from pandas.io.formats.style import Subset
import sys
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
x_train_full=pd.read_csv('train.csv',index_col='Id')
x_test_full=pd.read_csv('test.csv',index_col='Id')
pd.set_option('display.max_columns',None)
x_train_full.head()
```

```python
x_train_full.SalePrice.isnull().sum()
```

```python
y=x_train_full.SalePrice
x_train_full.drop(['SalePrice'],axis=1,inplace=True)
x_train_full,x_valid_full,y_train,y_test=train_test_split(x_train_full,y,train_size=0.8,test_size=0.2,random_state=0)
# "Cardinality" means the number of unique values in a column
# Select categorical columns with relatively low cardinality 
categorical_cols = [cname for cname in x_train_full.columns if x_train_full[cname].nunique() < 10 and x_train_full[cname].dtype == "object"]
# Select numerical columns
numerical_cols = [cname for cname in x_train_full.columns if x_train_full[cname].dtype in ['int64', 'float64']]
my_cols = categorical_cols + numerical_cols
x_train = x_train_full[my_cols].copy()
x_valid = x_valid_full[my_cols].copy()
X_test = x_test_full[my_cols].copy()
x_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Preprocessing for numerical data
numerical_transformer=SimpleImputer(strategy='constant')

# Preprocessing for categorical data
Preprocessor=ColumnTransformer(
    transfornmer=[
         ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

#Model selection
model=RandomForestRegressor(n_estimator=100,random_state=0)

# Bundle preprocessing and modeling code in a pipeline
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('model', model)
                     ])
clf.fit(X_train, y_train)
```
