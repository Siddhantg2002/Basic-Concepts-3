```python
import sys
import pandas as pd
import numpy as np
md=pd.read_csv('melb_data.csv')
pd.set_option('display.max_columns',None)
md.head()
```

```python
md.describe()
```

```python
md.shape
```

```python
md.columns
```

```python
md.isnull().sum()
```

```python
md=md.dropna(axis=0)
md.isnull().sum()
```

```python
md.shape
```

```python
y=md.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = md[melbourne_features]
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
md_model=DecisionTreeRegressor(random_state=1)
md_model.fit(X,y)
```

```python
predicted_home_prices = md_model.predict(X)
print(md_model.predict(X.head()))
```

```python
y.head()
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, predicted_home_prices)
```
